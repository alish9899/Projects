# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
# Importing File
df =  pd.read_excel(r"C:\Users\alish\Downloads\Customer Call List.xlsx")
df
```

```python
# Removing Duplicates

df = df.drop_duplicates()
df
```

```python
# Removing columns we dont need


df = df.drop(columns  = "Not_Useful_Column")
df
```

```python
# CustomerId looks fine, First_Name looks good too, lets have a look at Last_Name and clean up
#2 ways 
#df['Last_Name'] = df['Last_Name'].str.lstrip("...")
#df['Last_Name'] = df['Last_Name'].str.lstrip("/")
#df['Last_Name'] = df['Last_Name'].str.rstrip("_")

# or combined way
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df
```

```python
# Lets have a look at Phone_Number columns(cleaning up and standardize)


# lets get rid of all the non numeric values
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
# Now lets standardize the numbers 
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])


#We will need to convert the values to string 
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))


#Standardize the Phone numbers
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Now lets get rid of the remaining values
df["Phone_Number"] = df["Phone_Number"].str.replace('nan----','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na----','')
df
```

```python
# Now lets have a look at the Address column

df[["Street Address","State","Zip code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
# Lets now look at the remaining columns

df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
#df.replace('NaN','')

df = df.fillna('')
df
```

```python
# Now looking at the problems statement we need to provide the list with numbers one can call
# we will have to exclude people who dont want to be contacted and  also customer with no phone numbers


df1=df[(df['Do_Not_Contact']=='N') &(df['Phone_Number']!='')]
df1
```
